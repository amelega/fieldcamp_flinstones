```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def read_gp2(filename):
    eastings, northings = [], []
    with open(filename, 'r') as file:
        for line in file:
            parts = line.strip().split(',')  # Split by comma

            # Ensure there are enough columns before accessing indices
            if len(parts) >= 9:
                try:
                    easting = float(parts[8])  # 9th value (index 8)
                    northing = float(parts[6])  # 7th value (index 6)
                    eastings.append(easting)
                    northings.append(northing)
                except ValueError:
                    print(f"Skipping invalid line: {line.strip()}")  # Debugging info

    return eastings, northings
```

```python
plt.figure(figsize=(8, 6))
plt.xlabel("Easting (meters)")
plt.ylabel("Northing (meters)")
plt.title("GPS Path in Easting/Northing")

for i in range(1, 50):
    eastings, northings = read_gp2(f'../fieldcamp_flinstones/GPR/data/2024/Line{i}-ch2.gp2')
    plt.scatter(eastings, northings, marker='o', linestyle='-', color='blue')

plt.grid()
plt.show()
```
